```python
with WSDatabase() as db:
    result = db.query('SELECT average_life_expectancy_at_birth, alcohol_consumption FROM world_stats')

result.plot(kind='scatter', x='alcohol_consumption', y='average_life_expectancy_at_birth', alpha=0.5)
plt.title('Relationship between Alcohol Consumption and Average Life Expectancy')
plt.xlabel('Alcohol Consumption (per capita litres)')
plt.ylabel('Average Life Expectancy (years)')
plt.savefig('scatter.png')
```
